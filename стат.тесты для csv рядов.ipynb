<p><a href="https://statanaliz.info/statistica/proverka-gipotez/chto-takoe-proverka-statisticheskoj-gipotezy/" target="_blank">Переход на сайт statanaliz.info - выводы были подготовлены на основании этой статьи.</a></p> 

```python
# если исправишь α на 0.01, то напиши gr=2,58
# m - напиши число с которым хочешь среднее сравнить (истинное среднее генеральной совокупности (мат.ожидание))
def ztest1 (w, m = 5500, α = 0.05, gr=1.96):
        z=round((np.mean(w) - m) / (np.std(w)/ len(w)** 0.5),5)
        p=round((scipy.stats.norm.sf(abs(z))*2),5)
        
        print('  Среднее значение нашего ряда =', round(np.mean(w),2))
        print('  Предположим, что наше истинное среднее генеральной совокупности (мат.ожидание) равно', m,'.')
        print('  Допустим, что проверяемая нами гипотеза H0(ср.значение выборки = ср.значение ген.совокупности) верна.')
        print('  Альтернативная гипотеза - H1(ср.значение выборки ≠ ср.значение ген.совокупности).')
        print()
        print('  Проверку делаем на уровене значимости α =',α,',границы распределения поделены на:' )
        print('  • область принятия гипотезы H0: от -',gr,' до ',gr,';')
        print('  • области отклонения гипотезы H0: значения левее -',gr,' и правее',gr,').')      
        print()
        if abs(z)<1.96:
            if p>=α: 
                print('   ВЫВОД: критерий z=',z, ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          и p-value=',p ,' > 0,05 - гипотеза H0 не отклоняется :)')
            else: 
                print('   ВЫВОД: критерий z=',z, ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)     
                print('          но p-value=',p ,' < 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна:)')      
        if abs(z)>1.96:
            if p<α:
                print('   ВЫВОД: критерий z=', z , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)        
                print('          и p-value=',p ,'< 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна:)')      
            else:
                print('   ВЫВОД: критерий z=', z , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          и p-value=',p ,'> 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна ;)')      
print('---'* 40)
print(' '* 35, 'ОДНОВЫБОРОЧНЫЙ Z-ТЕСТ ДЛЯ РЯДА 1')
print('---'* 40)
ztest1(c) 
if j!=1:
    print()
    print('---'* 40)
    print(' '* 35, 'ОДНОВЫБОРОЧНЫЙ Z-ТЕСТ ДЛЯ РЯДА 2')
    print('---'* 40)
    ztest1(b)                
```

```python
# если исправишь α на 0.01, то напиши gr=2,58
def ztest2 (w1, w2, α = 0.05, gr=1.96):
        '''Объединенная стандартная ошибка'''
        oshibka2 = (np.std(w1, ddof=1)**2/len(w1) + np.std(w2, ddof=1)**2/len(w2))**0.5
        z2= round(((np.mean(w1) - np.mean(w2)) / oshibka2),5)
        p2 = round((scipy.stats.norm.sf(abs(z2))*2),5)
        
        
        print('  Ряд 1: среднее значение  =',round(np.mean(w1),2), ', количество наблюдений =',len(w1),', дисперсия=',round(np.var(w1,ddof=1),2))
        print('  Ряд 2: среднее значение  =',round(np.mean(w2),2), ', количество наблюдений =',len(w2),', дисперсия=',round(np.var(w2,ddof=1),2))
        print()
        print('  Допустим, что проверяемая нами гипотеза H0(средние значения двух рядов равны) верна.')
        print('  Альтернативная гипотеза - H1(средние значения двух рядов не равны).')
        print()
        print('  Проверку делаем на уровене значимости α =',α,',границы распределения поделены на:' )
        print('  • область принятия гипотезы H0: от -',gr,' до ',gr,';')
        print('  • области отклонения гипотезы H0: значения левее -',gr,' и правее',gr,').')      
        print()
        if abs(z2)<1.96:
            if p2>=α: 
                print('   ВЫВОД: критерий z=',z2, ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          и p-value=',p2 ,' > 0,05 - гипотеза H0 не отклоняется :)')
            else: 
                print('   ВЫВОД: критерий z=',z2, ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)     
                print('          но p-value=',p2 ,' < 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна :)')      
        if abs(z2)>1.96:
            if p2<α:
                print('   ВЫВОД: критерий z=', z2 , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)        
                print('          и p-value=',p2 ,'< 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 вернао:)')      
            else:
                print('   ВЫВОД: критерий z=', z2 , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          и p-value=',p2 ,'> 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна ;)')      
print('---'* 40)
print(' '* 35, 'ДВУХВЫБОРОЧНЫЙ Z-ТЕСТ ДЛЯ СРЕДНИХ РЯДА 1 И РЯДА 2')
print('---'* 40)
ztest2(c,b) 
```

```python
# если исправишь α на 0.01, то напиши gr=2,58
def ttest (w3, w4, α = 0.05, gr=1.96):
        t,p3 = scipy.stats.ttest_ind(w3, w4)
                        
        print('  Ряд 1: среднее значение  =',round(np.mean(w3),2), ', количество наблюдений =',len(w3),', дисперсия=',round(np.var(w3,ddof=1),2))
        print('  Ряд 2: среднее значение  =',round(np.mean(w4),2), ', количество наблюдений =',len(w4),', дисперсия=',round(np.var(w4,ddof=1),2))
        print()
        print('  Допустим, что проверяемая нами гипотеза H0(средние значения двух рядов равны) верна.')
        print('  Альтернативная гипотеза - H1(средние значения двух рядов не равны).')
        print()
        print('  Проверку делаем на уровене значимости α =',α,',границы распределения поделены на:' )
        print('  • область принятия гипотезы H0: от -',gr,' до ',gr,';')
        print('  • области отклонения гипотезы H0: значения левее -',gr,' и правее',gr,').')      
        print()
        if abs(t)<1.96:
            if p3>=α: 
                print('   ВЫВОД: критерий t=',round(t,5), ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          и p-value=',round(p3,5) ,' > 0,05 - гипотеза H0 не отклоняется :)')
            else: 
                print('   ВЫВОД: критерий t=',round(t,5), ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)     
                print('          но p-value=',round(p3,5) ,' < 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна:)')      
        if abs(t)>1.96:
            if p3<α:
                print('   ВЫВОД: критерий t=', round(t,5) , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)        
                print('          и p-value=',round(p3,5) ,'< 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна:)')      
            else:
                print('   ВЫВОД: критерий t=', round(t,5) , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          и p-value=',round(p3,5) ,'> 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна ;)')      
print('---'* 40)
print(' '* 35, 'T-ТЕСТ НА РАВЕНСТВО СРЕДНИХ РЯДА 1 И РЯДА 2')
print('---'* 40)
ttest(c,b)
```

F-ТЕСТ НА РАВЕНСТВО ДИСПЕРСИЙ </font>

```python
# если исправишь α на 0.01, то напиши gr=2,58
def ftest (w5, w6, α = 0.05, gr=1.96):
        F = round(max(np.var(w5,ddof=1), np.var(w6,ddof=1))/min(np.var(w5,ddof=1), np.var(w6,ddof=1)),5)
        fdistribution = stats.f(len(w5)-1,len(w6)-1) 
        p4 = round((1-fdistribution.cdf(F)),5)
        f_critical = round(fdistribution.ppf(1-α),5)
        
        print('  Ряд 1: среднее значение  =',round(np.mean(w5),2), ', количество наблюдений =',len(w5),', дисперсия=',round(np.var(w5,ddof=1),2))
        print('  Ряд 2: среднее значение  =',round(np.mean(w6),2), ', количество наблюдений =',len(w6),', дисперсия=',round(np.var(w6,ddof=1),2))
        print()
        print('  Допустим, что проверяемая нами гипотеза H0(дисперсии двух рядов равны) верна.')
        print('  Альтернативная гипотеза - H1(дисперсии двух рядов не равны).')
        print()
        print('  Проверку делаем на уровене значимости α =',α,',границы распределения поделены на:' )
        print('  • область принятия гипотезы H0: от -',gr,' до ',gr,';')
        print('  • области отклонения гипотезы H0: значения левее -',gr,' и правее',gr,').')      
        print()
        if abs(F)<1.96:
            if p4>=α: 
                print('   ВЫВОД: критерий F=',F, ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          F критическое одностороннее =', f_critical, ', p-value=',p4 ,' > 0,05 - гипотеза H0 не отклоняется :)')
            else: 
                print('   ВЫВОД: критерий F=',F, ', он находится в пределах области принятия гипотезы H0 от -',gr,' до ',gr)     
                print('          F критическое одностороннее =', f_critical, ', p-value=',p4 ,' < 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна:)')      
        if abs(F)>1.96:
            if p4<α:
                print('   ВЫВОД: критерий F=', F , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)        
                print('          F критическое одностороннее =', f_critical, ', p-value=',p4 ,'< 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна:)')      
            else:
                print('   ВЫВОД: критерий F=', F , ', он находится за пределами области принятия гипотезы H0 от -',gr,' до ',gr)
                print('          F критическое одностороннее =', f_critical, ', p-value=',p4 ,'> 0,05 - гипотеза H0 отклоняется как ложная (точнее, маловероятная),')
                print('          предположительно альтернативная гипотеза H1 верна ;)')      
print('---'* 40)
print(' '* 35, 'F-ТЕСТ НА РАВЕНСТВО ДИСПЕРСИЙ РЯДА 1 И РЯДА 2')
print('---'* 40)
ftest(c,b)
```
